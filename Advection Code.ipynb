```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ngrids = 10
L = 1.0
dx = L / ngrids
dt = 0.01
c =1
def square_wave(x):
    if 0.25 <= x <= 0.75:
        return 1.0
    else:
        return 0.0
u = np.array([square_wave(x) for x in np.linspace(0, L, ngrids)])
print(u)    
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.linspace(0, L, ngrids), u)
plt.xlabel("grid point")
plt.ylabel("u")
plt.show()
```

### Forward difference

u(i,t+1) = u(i,t) - c*dt/dx*( u(i+1,t) - u(i,t) )

```python
def forward_differece(u, L, c, dt, dx, ngrids):
    u_old = u
    plt.figure(figsize=(6,4))
    plt.plot(np.linspace(0, L, ngrids), u_old, label="old", color="b")
    for i in range(len(u)):
        if i < len(u)-1:
            u[i] = u_old[i] - c* dt/dx * (u_old[i+1] - u_old[i])
    plt.plot(np.linspace(0, L, ngrids), u, label="new", color="r")
    plt.xlabel("grid point")
    plt.ylabel("u")
    plt.legend()
    plt.show()               
```

```python
ngrids = 10
L = 1.0
dx = L / ngrids
dt = 0.01
c =1
u = np.array([square_wave(x) for x in np.linspace(0, L, ngrids)])
forward_differece(u, L, c, dt, dx, ngrids)
```

### Backward difference

u(i,t+1) = u(i,t) - c*dt/dx*( u(i,t) - u(i-1,t) )

```python
def backward_differece(u, L, c, dt, dx, ngrids):
    u_old = u
    plt.figure(figsize=(6,4))
    plt.plot(np.linspace(0, L, ngrids), u_old, label="old", color="b")
    for i in range(len(u)):
        if i < len(u)-1:
            u[i] = u_old[i] - c* dt/dx * (u_old[i] - u_old[i-1])
    plt.plot(np.linspace(0, L, ngrids), u, label="new", color="r")
    plt.xlabel("grid point")
    plt.ylabel("u")
    plt.legend()
    plt.show()               
```

```python
ngrids = 10
L = 1.0
dx = L / ngrids
dt = 0.01
c =1
u = np.array([square_wave(x) for x in np.linspace(0, L, ngrids)])
backward_differece(u, L, c, dt, dx, ngrids)
```

### Central difference

u(i,t+1) = u(i,t) - c*dt/dx*( u(i+1,t) - u(i-1,t) )/2

```python
def central_differece(u, L, c, dt, dx, ngrids):
    u_old = u
    plt.figure(figsize=(6,4))
    plt.plot(np.linspace(0, L, ngrids), u_old, label="old", color="b")
    for i in range(len(u)):
        if i < len(u)-1:
            u[i] = u_old[i] - c* dt/dx * (u_old[i+1] - u_old[i-1])/2
    plt.plot(np.linspace(0, L, ngrids), u, label="new", color="r")
    plt.xlabel("grid point")
    plt.ylabel("u")
    plt.legend()
    plt.show()               
```

```python
ngrids = 10
L = 1.0
dx = L / ngrids
dt = 0.01
c =1
u = np.array([square_wave(x) for x in np.linspace(0, L, ngrids)])
central_differece(u, L, c, dt, dx, ngrids)
```
